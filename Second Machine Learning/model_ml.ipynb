```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import OrdinalEncoder

# membaca datanya
df = pd.read_csv('train.csv')
```

```python
# klasifikasikan datanya
# top_features = [
#     "OverallQual",
#     "GrLivArea",
#     "GarageCars",
#     "GarageArea",
#     "TotalBsmtSF",
#     "1stFlrSF",
#     "FullBath",
#     "TotRmsAbvGrd",
#     "YearBuilt",
#     "YearRemodAdd",
#     "GarageYrBlt",
#     "MasVnrArea",
#     "Fireplaces",
#     "BsmtFinSF1",
#     "LotFrontage",
#     "WoodDeckSF"
# ]

x = df
y = df.SalePrice
```

```python
# pisahkan
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
# memulai debugging
s = (df.dtypes == 'str') | (df.dtypes == 'object')
object_cols = list(s[s].index)
print(object_cols)
```

```python
# membuat fungsi untuk mengetahui
model_3 = RandomForestRegressor(
    n_estimators=100,
    criterion='squared_error',
    random_state=0
)

def score_get_mae(x_t=X_train, x_v=X_test, y_t=y_train, y_v=y_test):
    model_3.fit(x_t, y_v)
    prediction = model_3.predict(x_v)
    my_mae = mean_absolute_error(y_v, prediction)

    return my_mae

# berhasil mendefinisikan                                                                       
```

```python
# karena sudah dapet jadi tinggal masukan aja ke encoding
# approach 2
label_x_train = X_train.copy()
label_x_test = X_test.copy()

# mulai lakukan ordinal encoding
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
label_x_train[object_cols] = ordinal_encoder.fit_transform(X_train[object_cols])
label_x_test[object_cols] = ordinal_encoder.transform(X_test[object_cols])

print("MAE yang didapatkan dengan approach 2 adalah")
print(score_get_mae(label_x_train, label_x_test, y_train, y_test))
```
