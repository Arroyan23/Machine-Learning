```python
# mencari datasets
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

file_directory = 'train.csv'

df = pd.read_csv(file_directory)

df.columns
```

```python
# definisikan top features nya
top_features = [
    "OverallQual",
    "GrLivArea",
    "GarageCars",
    "GarageArea",
    "TotalBsmtSF",
    "1stFlrSF",
    "FullBath",
    "TotRmsAbvGrd",
    "YearBuilt",
    "YearRemodAdd",
    "GarageYrBlt",
    "MasVnrArea",
    "Fireplaces",
    "BsmtFinSF1",
    "LotFrontage",
    "WoodDeckSF"
]

# masukkan ke x dan y
x = df[top_features]
y = df.SalePrice

x.head()
```

```python
# memisahkan train test
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.2, random_state=1)
print("Variabel berhasil dipisahkan...")
```

```python
# gunakan random forest dan buat 5 model untuk menguji
model_1 = RandomForestRegressor(
    n_estimators=50, 
    random_state=0
    )
model_2 = RandomForestRegressor(
    n_estimators=100, 
    min_samples_leaf=5, 
    random_state=0
    )
model_3 = RandomForestRegressor(
    n_estimators=100, 
    criterion='squared_error', 
    random_state=0
    )
model_4 = RandomForestRegressor(
    n_estimators=100,
    criterion='absolute_error', 
    random_state=0
    )
model_5 = RandomForestRegressor(
    n_estimators=200, 
    max_depth=7, 
    random_state=0
    )
model_6 = RandomForestRegressor(
    n_estimators=100,
    max_features=0.5,   # atau 'sqrt', 'log2', atau angka int
    random_state=0
)
model_7 = RandomForestRegressor(
    n_estimators=100,
    min_samples_split=10,
    random_state=0
)
model_8 = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    min_samples_leaf=3,
    random_state=0
)

models = [model_1, model_2, model_3, model_4, model_5, model_6, model_7, model_8]
```

```python
# looping untuk mengetahui model mana yang bagus
# buat fungsi untuk mendapatkan score
def score_model(how_model, t_x=train_x, v_x=test_x, t_y=train_y, v_y=test_y):
    how_model.fit(t_x,t_y)
    predicts = how_model.predict(v_x)
    my_mae = mean_absolute_error(v_y, predicts)

    return my_mae

for i in range(0, len(models)):
    mae = score_model(models[i])
    print(f"Dengan model : %d \t\t MAE score : {mae:,.0f}" %(i+1))
```

```python
best_model = model_3
best_model.fit(x, y)
print("successfully learned...")
```

```python
# masukkan ke dalam data test
dft = pd.read_csv('test.csv')

# ambil parameter yang diperlukan
x_evaluation = dft[top_features]

x_evaluation.head()
```

```python
# mulai prediksi model
predicted = best_model.predict(x_evaluation)
print("Successfully predicted...")
print(predicted)
```

```python
# buat excel submission
submission = pd.DataFrame({
    'Id' : dft['Id'],
    'SalePrice' : predicted
})

submission.to_csv("submission.csv", index=False)
print(submission.head())
```
