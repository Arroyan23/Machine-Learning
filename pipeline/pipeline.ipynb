```python
# testing 123
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer

# mengambil datanya
df = pd.read_csv("train.csv")
dft = pd.read_csv("test.csv")

print("berhasil mengambil datanya....")
```

```python
# memisahkan dengan train test split
X = df.drop(['SalePrice'], axis=1)
y = df.SalePrice
X_train_full, X_test_full, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# display(X.head())
display(X_train_full.head())
```

```python
# memisahkan data category dan data numerical
object_cols = X_train_full.select_dtypes(include='object')
category_cols = [cname for cname in object_cols if
                  X_train_full[cname].nunique() < 10]

# print(category_cols)
# ambil numerical cols
numerical_cols = [cname for cname in X_train_full if X_train_full[cname].dtypes in ['int64', 'float64']]
print("Numerical cols nya adalah")
# print(numerical_cols)
# berhasil mendapatkan categorical dan numerical
```

```python
# mengcopy filenya dan mengidentifikasi dengan full
all_cols = category_cols + numerical_cols
X_train = X_train_full[all_cols].copy()
X_valid = X_test_full[all_cols].copy()
X_test = dft[all_cols].copy()

X_train.head()
```

```python
# selanjutnya lakukan pipeline model
# preprocessing menggunakan pipeline
# preprocessing numerical data
numerical_transformer = SimpleImputer(strategy='constant')

# jadi untuk numerical cukup di simple imputer saja
categorical_transformer = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(
        handle_unknown='ignore'
    ))
])

# berhasil membuat modelnya
# bundle processing for numerical and categorical
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('category', categorical_transformer, category_cols)
    ]
)

print("Berhasil membuat mesin preprocessing")
```

```python
# membuat mesin model
model = RandomForestRegressor(n_estimators=200, random_state=42)
```

```python
# membuat pipeline modelnya
from sklearn.metrics import mean_absolute_error
my_pipe = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('model', model)
])

# preprocess training data dan fit modelnya
my_pipe.fit(X_train, y_train)

# melakukan prediksi
prediction_train = my_pipe.predict(X_valid)
print("berhasil melakukan prediksi")
score_mae = mean_absolute_error(y_test, prediction_train)
print(f"Score MAE nya adalah : {score_mae}")
```
