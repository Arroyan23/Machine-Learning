```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

# load data dengan pandas
file_directory = 'melb_data_house.csv'
read_data = pd.read_csv(file_directory)

read_data.describe()
```

```python
# identifikasi variabel 
read_data.columns
```

```python
data_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
# masukkan ke dalam x
x = read_data[data_features]

# definisikan y dengan harga
y = read_data.Price

x.head()
```

```python
# pisahkan dengan train dan test
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

# buat modelnya
housing_model = DecisionTreeRegressor()
housing_model.fit(x_train, y_train)

# lakukan prediksi dengan y test
house_price_prediction = housing_model.predict(x_test)
# print(house_price_prediction)
# print("Perbandingan dengan nilai asli")
# print(y_test.head(10))

# hitung selisih dari prediksi dengan nilai asli
difference = y_test - house_price_prediction

comparison = pd.DataFrame({
    "Actual": y_test,
    "Predicted": house_price_prediction,
    "Sum" : difference
})
print(comparison.head(10))

mae_values = mean_absolute_error(y_test, house_price_prediction)

print(f"Mean Absolute Error: {mae_values}")
```
