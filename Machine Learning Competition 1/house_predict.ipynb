```python
# import yang dibutuhkan terlebih dahulu
import pandas as pd
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# baca filenya
read_file = 'train.csv'
home_file = pd.read_csv(read_file)
# tes dengan membaca filenya

home_file.describe()
```

```python
# liat apa saja kolom yang disediakan oleh data tersebut
home_file.columns
```

```python
# pisahkan hanya yang berguna untuk prediksi sale pricenya
top_features = [
    "OverallQual",
    "GrLivArea",
    "GarageCars",
    "GarageArea",
    "TotalBsmtSF",
    "1stFlrSF",
    "FullBath",
    "TotRmsAbvGrd",
    "YearBuilt",
    "YearRemodAdd",
    "GarageYrBlt",
    "MasVnrArea",
    "Fireplaces",
    "BsmtFinSF1",
    "LotFrontage",
    "WoodDeckSF"
]

x = home_file[top_features]

y = home_file.SalePrice

x.head()
```

```python
# pisahkan variabel train dan test dalam data itu sendiri
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state=0)
print("Separating success...")
```

```python
# buat model decision tree
house_model_price = DecisionTreeRegressor(random_state=1)
house_model_price.fit(train_x, train_y)

# prediksi hasil
house_saleprice_predict = house_model_price.predict(test_x)
print("Prediksi Berhasil...")

print(house_saleprice_predict)
```

```python
# mengetahui berapa errornya
rmse = root_mean_squared_error(test_y, house_saleprice_predict)
print(f"Nilai rmse nya adalah : {rmse:,.0f}")
```

```python
# mendapatkan rmse nya
def get_rmse(max_leaf_nodes, x_train, x_test, y_train, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=42)
    model.fit(x_train, y_train)
    predict_result = model.predict(x_test)
    rmse = root_mean_squared_error(y_test, predict_result)

    return rmse
```

```python
# menggunakan max leaf sebagai penilaian
for max_leaf_nodes in [5, 25, 50, 100, 250, 500, 1000]:
    my_rmse = get_rmse(max_leaf_nodes, train_x, test_x, train_y, test_y)
    print(f"Max Leaf Nodes : {max_leaf_nodes} \t\t RMSE : {my_rmse:,.0f}")

best_max_leaf_nodes = 100
```

```python
# mencoba menggunakan random forest regressor
model_regressor = RandomForestRegressor(random_state=1)
model_regressor.fit(train_x, train_y)

predict_with_random_forest = model_regressor.predict(test_x)

# hitung rmse nya
rmse_random_forest = root_mean_squared_error(test_y, predict_with_random_forest)

print(f"Hasil rmse dari random forest regressor adalah : {rmse_random_forest:,.0f}")
# karena rendah maka memakai random forest regressor
```

```python
# memulai prediksi dengan data test
test_data = 'test.csv'
read_test_data = pd.read_csv(test_data)

read_test_data.head()
```

```python
# ambil data features dari data test
x_compe_data = read_test_data[top_features]
# mulai prediksi
final_predict = model_regressor.predict(x_compe_data)
print(f"Rata rata hasil prediksi adalah : {final_predict.mean():,.3f}")
print("Prediksi berhasil....")
```

```python
# masukkan ke dalam file submission
# Ganti 'id' di bawah sesuai nama kolom asli di dataset kamu (misal: 'Id', 'ID', 'user_id')
submission = pd.DataFrame({
    'Id': read_test_data['Id'],
    'SalePrice': final_predict
})

submission.to_csv("submission.csv", index=False)
print(submission.head())
```
